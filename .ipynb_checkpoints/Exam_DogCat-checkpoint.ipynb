```python
# インポート
import numpy as np
import pandas as pd
from PIL import Image
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 訓練データの読み込み
train_data = pd.read_csv("dc_photos/train/train_data.csv")
train_data[::25]
```

```python
# テストデータの読み込み
test_data = pd.read_csv("dc_photos/test/test_data.csv")
test_data[::25]
```

```python
# 訓練画像データをそのままの表示
sample_img1 = Image.open("dc_photos/train/cat-001.jpg")
plt.imshow(sample_img1)
```

```python
# 訓練画像データをグレースケールに変換して表示
sample_img1_gray = sample_img1.convert("L")
plt.imshow(sample_img1_gray, cmap="gray")
```

```python
# テスト画像データをそのまま表示
sample_img2 = Image.open("dc_photos/test/cat-151.jpg")
plt.imshow(sample_img2)
```

```python
# テスト画像データをグレースケールで表示
sample_img2_gray = sample_img2.convert("L")
plt.imshow(sample_img2_gray, cmap="gray")
```

```python
# グレースケール画像のサイズ確認
sample_img1_gray_array = np.array(sample_img1_gray)
sample_img1_gray_array.shape
```

```python
# 訓練用データの読み込みと前処理
#  -> 前処理として元データと180度回転の２パターン作成

# ndarrayのデータを保管するための領域確保
img_size = 75 * 75
train_len = len(train_data)
X_train = np.empty((train_len * 2, img_size), dtype=np.uint8)
y_train = np.empty(train_len * 2, dtype=np.uint8)

# 画像ひとつひとつについて繰り返し処理
for i in range(train_len):
    
    # 基画像をndarrayとして読み込んで訓練データに追加
    name = train_data.loc[i, "File name"]
    train_img = Image.open(f"dc_photos/train/{name}.jpg").convert("L")
    train_img = np.array(train_img)
    train_img_f = train_img.flatten()
    X_train[i] = train_img_f
    y_train[i] = train_data.loc[i, "DC"]
    
    # 180度回転させたものを訓練データに追加
    train_img = np.rot90(train_img, 2)
    train_img_f = train_img.flatten()
    X_train[i + train_len] = train_img_f
    y_train[i + train_len] = train_data.loc[i, "DC"]
```

```python
# テスト用データの読み込み

# ndarrayのデータを保管するための領域確保
img_size = 75 * 75
test_len = len(test_data)
X_test = np.empty((test_len, img_size), dtype=np.uint8)
y_test = np.empty(test_len, dtype=np.uint8)

# 画像ひとつひとつについて繰り返し処理
for i in range(test_len):
    
    # ndarrayとして読み込んでテストデータに追加
    name = test_data.loc[i, "File name"]
    test_img = Image.open(f"dc_photos/test/{name}.jpg").convert("L")
    test_img = np.array(test_img)
    test_img_f = test_img.flatten()
    X_test[i] = test_img_f
    y_test[i] = test_data.loc[i, "DC"]
```

```python
# モデルを使って学習

# 分類器の作成
classifier = SVC(kernel="linear", gamma="scale", random_state=1)
classifier.fit(X_train, y_train)
```

```python
# 期待する性能が出たか評価
y_pred = classifier.predict(X_test)
y_pred
```

```python
# 正解の表示
y_test
```

```python
# 混合行列で正答数の表示
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
# 正答率の表示
print(metrics.classification_report(y_test, y_pred))
```
