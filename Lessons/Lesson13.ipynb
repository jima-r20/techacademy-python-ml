```python
%matplotlib inline
# ↑　JupyterLabのノートブックを利用してブラウザにグラフを表示する場合には必須
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.random.rand(5)
y *=5
plt.plot(y)
```

```python
# 横軸のデータの用意
x = np.arange(1, 11, 2)
y = np.random.rand(5)
y *= 5
plt.plot(x, y)  # plt.plot(横軸の配列, 縦軸の配列)
```

```python
# グラフのスタイル変更 -> plot()に引数を追加する
x = np.arange(1, 11, 2)
y = np.random.rand(5)
y *= 5
plt.plot(x, y, linestyle = "--", marker = "^", color = "r")
```

```python
# グラフの部品追加 -> タイトルやラベルなど
x = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
y = np.array([400, 300, 350, 450, 400, 600, 500, 450])

plt.title("Year - Count")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)

plt.plot(x, y)
```

```python
# 棒グラフ -> 縦の棒グラフ bar(), 横の棒グラフ barh()
year = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
count = np.array([400, 300, 350, 450, 400, 600, 500, 450])

plt.bar(year, count)
```

```python
plt.barh(year, count)
```

```python
# 散布図 -> scatter(横軸, 縦軸)
english = np.array([70, 95, 83, 100, 72, 71, 75, 85, 67, 75])
math = np.array([90, 82, 100, 95, 50, 81, 52, 56, 67, 71])

plt.scatter(english, math)
```

```python
# 円グラフ -> pie()
fluits = np.array(["peach", "orange", "grape", "banana", "apple"])
count = np.array([12, 20, 34, 36, 47])

plt.pie(count, labels = fluits, autopct = "%.1f%%", startangle = 90)
#  -> label:各要素のラベル, autopct:データの書式設定, startangle:円グラフを開始する角度（基準を３時の方向として、反時計回りに何度か指定）
```

```python
# 複数のグラフを一度に表示

# 一つのグラフ領域に２つの線グラフ
y1 = np.random.rand(11)
y2 = np.random.rand(11)

plt.plot(y1, color = "blue", label = "y1")
plt.plot(y2, color = "red", label = "y2")
plt.legend(loc = "upper left")  # legend(): 凡例を表示するためのメソッド -> plot()でlabelを追加する必要あり。locは表示位置
```

```python
# 複数のグラフを並べて表示
y1 = np.random.rand(11)
y2 = np.random.rand(11)

fig = plt.figure()  # matplotlib.figure.Figure型のオブジェクト生成

sp1 = fig.add_subplot(1, 2, 1)  # 上記で生成したオブジェクトにadd_subplotで絵画領域を指定し、グラフ画像追加。引数は(行数, 列数, 何番目)
sp1.plot(y1)

sp2 = fig.add_subplot(1, 2, 2)
sp2.plot(y2)
```

```python
# 近似直線

# 最小二乗法による近似直線
x = np.array([20, 30, 40, 50, 60])
y = np.array([30, 38, 41, 49, 62])

p = np.polyfit(x, y, 1)  #最小二乗法による近似直線の式(y = ax + b)の導出(NumPy)。引数(xに該当するデータ, yに該当するデータ, 1) -> 1は一次式を示す
f = np.poly1d(p)  # 一次関数の式のオブジェクト生成

plt.scatter(x, y)
plt.plot(x, f(x))
```
